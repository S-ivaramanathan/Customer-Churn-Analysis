```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df["Churn"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Churn"].value_counts().plot(kind="bar")

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
```

```python
pd.crosstab(df["gender"], df["Churn"])
```

```python
pd.crosstab(df["gender"], df["Churn"]).plot(kind="bar")

plt.title("Gender vs Churn")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

```python
pd.crosstab(df["Contract"], df["Churn"])
```

```python
pd.crosstab(df["Contract"], df["Churn"]).plot(kind="bar", figsize=(8,5))

plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df.groupby("Churn")["MonthlyCharges"].mean()
```

```python
import matplotlib.pyplot as plt

df.boxplot(column="MonthlyCharges", by="Churn")

plt.title("Monthly Charges vs Churn")
plt.suptitle("")   # Removes the extra default title
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")
plt.show()
```

```python
df.groupby("Churn")["tenure"].mean()
```

```python
df.boxplot(column="tenure", by="Churn")

plt.title("Tenure vs Churn")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("Tenure (Months)")
plt.show()
```

```python
pd.crosstab(df["PaymentMethod"], df["Churn"])
```

```python
pd.crosstab(df["PaymentMethod"], df["Churn"]).plot(kind="bar", figsize=(10,5))

plt.title("Payment Method vs Churn")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.xticks(rotation=30)
plt.show()
```

```python
pd.crosstab(df["InternetService"], df["Churn"])
```

```python
pd.crosstab(df["InternetService"], df["Churn"]).plot(kind="bar", figsize=(8,5))

plt.title("Internet Service vs Churn")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")
plt.show()
```
